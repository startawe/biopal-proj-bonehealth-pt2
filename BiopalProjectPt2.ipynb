# Background and Rationale

***4.1 Health Issue or Knowledge Gap***

Bone health is a major concern in aging populations due to the increasing risk of osteoporosis and related fractures. While low bone mineral density (BMD) is a recognized risk factor for fragility fractures, the relationship between obesity and bone health remains complex and underexplored. Traditionally, increased body weight was assumed to be protective against low BMD due to the mechanical loading of bones. However, recent evidence suggests that excess adiposity may exert negative influences on bone metabolism by increasing systemic inflammation, promoting insulin resistance, and altering the hormonal regulation of calcium and vitamin D.

There is a lack of comprehensive analysis that integrates bone mineral density (BMD), serum vitamin D, and calcium levels within the context of obesity and modifiable lifestyle factors (e.g., smoking, alcohol use, physical activity). This gap limits our understanding of how behavioral and demographic factors mediate the obesity–bone health relationship.


***4.2 Importance of the research question in Public Health and Biomedical Science***

Conflicting evidence in current research creates uncertainty for clinicians and public health professionals regarding bone health management in obese individuals. The assumption that higher body weight always equates to stronger bones may lead to missed opportunities for early detection and intervention in those at risk. At the same time, insufficient understanding of the biochemical and behavioral factors that modify this relationship hinders effective policy and clinical recommendations.

This research will help clarify the complex relationships between obesity, micronutrient status, behavior, and bone health, providing actionable insights to reduce the burden of osteoporosis and fractures in aging adults—improving quality of life and reducing healthcare costs.

```python
import pandas as pd
import pyreadstat

# ----------------------------
# Step 1: Define file groups
# ----------------------------
dxxfem_files = [
    "DXXFEM_D 2005-2006.xpt", "DXXFEM_E 2007-2008.xpt", "DXXFEM_F 2009-2010.xpt",
    "DXXFEM_H 2013-2014.xpt", "DXXFEM_J 2017-2018.xpt", "P_DXXFEM 2017-2020.xpt"
]

demo_files = [
    "DEMO_D 2005-2006.xpt", "DEMO_E 2007-2008.xpt", "DEMO_F 2009-2010.xpt",
    "DEMO_H 2013-2014.xpt", "DEMO_J 2017-2018.xpt", "P_DEMO 2017-2020.xpt"
]

bmx_files = [
    "BMX_D 2005-2006.xpt", "BMX_E 2007-2008.xpt", "BMX_F 2009-2010.xpt",
    "BMX_H 2013-2014.xpt", "BMX_J 2017-2018.xpt", "P_BMX 2017-2020.xpt"
]

smq_files = [
    "SMQ_D 2005-2006.xpt", "SMQ_E 2007-2008.xpt", "SMQ_F 2009-2010.xpt",
    "SMQ_H 2013-2014.xpt", "SMQ_J 2017-2018.xpt", "P_SMQ 2017-2020.xpt"
]

alq_files = [
    "ALQ_D 2005-2006.xpt", "ALQ_E 2007-2008.xpt", "ALQ_F 2009-2010.xpt",
    "ALQ_H 2013-2014.xpt"
]

paq_files = [
    "PAQ_E 2007-2008.xpt", "PAQ_F 2009-2010.xpt", "PAQ_H 2013-2014.xpt",
    "PAQ_J 2017-2018.xpt", "P_PAQ 2017-2020.xpt"
]

# ----------------------------
# Step 2: Helper function to read and concatenate files
# ----------------------------
def read_and_concat(files, cols_to_keep):
    dfs = []
    for f in files:
        try:
            df = pd.read_sas(f, format='xport', encoding='latin1')
            cols_present = [col for col in cols_to_keep if col in df.columns]
            df_filtered = df[cols_present]
            dfs.append(df_filtered)
        except Exception as e:
            print(f"❌ Failed to read {f}: {e}")
    if dfs:
        return pd.concat(dfs, ignore_index=True)
    else:
        return pd.DataFrame(columns=cols_to_keep)

# ----------------------------
# Step 3: Load datasets
# ----------------------------
df_dxxfem = read_and_concat(dxxfem_files, ['SEQN', 'DXXOFBMD'])
df_demo   = read_and_concat(demo_files, ['SEQN', 'SDDSRVYR', 'RIDAGEYR', 'RIAGENDR'])
df_bmx    = read_and_concat(bmx_files, ['SEQN', 'BMXBMI', 'BMXWAIST'])
df_smq    = read_and_concat(smq_files, ['SEQN', 'SMQ020'])
df_alq    = read_and_concat(alq_files, ['SEQN', 'ALQ101'])
df_paq    = read_and_concat(paq_files, ['SEQN', 'PAQ605', 'PAQ650'])

# ----------------------------
# Step 4: Merge all data on SEQN
# ----------------------------
df_merged = df_dxxfem.merge(df_demo, on='SEQN', how='left') \
                     .merge(df_bmx, on='SEQN', how='left') \
                     .merge(df_smq, on='SEQN', how='left') \
                     .merge(df_alq, on='SEQN', how='left') \
                     .merge(df_paq, on='SEQN', how='left')

# ----------------------------
# Step 5: Drop rows missing BMD values
# ----------------------------
df_merged = df_merged[df_merged['DXXOFBMD'].notna()].copy()

# ----------------------------
# Step 6: Reset index
# ----------------------------
df_merged.reset_index(drop=True, inplace=True)

# ----------------------------
# Step 7: Replace missing values with "N/A"
# ----------------------------
# Reverting fillna for columns that should remain numeric for pyreadstat
# df_merged = df_merged.fillna("N/A") # Commenting this out

# ----------------------------
# Step 8: Convert coded columns to readable labels
# ----------------------------
gender_map = {1: "Male", 2: "Female"}
activity_map = {
    1: "Yes",
    2: "No",
    7: "Refused",
    9: "Don't know"
}

df_merged["RIAGENDR"] = df_merged["RIAGENDR"].replace(gender_map)
# Keep PAQ605 and PAQ650 as original values, as they are used for filtering later
# df_merged["PAQ605"] = df_merged["PAQ605"].replace(activity_map)
# df_merged["PAQ650"] = df_merged["PAQ650"].replace(activity_map)


# ----------------------------
# Step 9: Preview the final dataset
# ----------------------------
print("✅ Final dataset shape:", df_merged.shape)
print(df_merged.head())

# ----------------------------
# Step 10: Convert potentially mixed-type columns to string before saving
# This step is no longer needed after removing the fillna("N/A")
# for col in ['SMQ020', 'ALQ101']:
#     if col in df_merged.columns:
#         df_merged[col] = df_merged[col].astype(str)


# ----------------------------
# Step 11: Save merged dataset to .xpt
# ----------------------------
pyreadstat.write_xport(df_merged, "finalbmd_no_rxq_osq.xpt", table_name="NHANES_DATA")
```

```python
# Import necessary libraries
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# Ignore warnings from statsmodels
warnings.filterwarnings("ignore", category=UserWarning)

# ----------------------------------------------------------------------------
# Step 1: Define variables
# ----------------------------------------------------------------------------
dependent_variable = 'DXXOFBMD'
categorical_independent_variables = ['RIAGENDR', 'PAQ605', 'PAQ650', 'ALQ101', 'SMQ020']
continuous_covariates = ['RIDAGEYR', 'BMXBMI', 'BMXWAIST']

# ----------------------------------------------------------------------------
# Step 2: Clean data - Drop rows with missing values in dependent variable and main covariates
# ----------------------------------------------------------------------------
# Ensure continuous covariates are numeric and convert 'N/A' to NaN
for col in continuous_covariates:
    df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce')

# Drop rows with missing values in the dependent variable and continuous covariates
df_cleaned = df_merged.dropna(subset=[dependent_variable] + continuous_covariates).copy()

# ----------------------------------------------------------------------------
# Step 3: Convert all categorical independent variables to category dtype
# ----------------------------------------------------------------------------
for col in categorical_independent_variables:
    df_cleaned[col] = df_cleaned[col].astype('category')

# ----------------------------------------------------------------------------
# Step 4: Loop through all categorical independent variables and run ANOVA/ANCOVA
# ----------------------------------------------------------------------------
analysis_results = {}

for categorical_variable in categorical_independent_variables:
    print(f"\nAnalyzing {categorical_variable} vs. {dependent_variable} (controlling for {', '.join(continuous_covariates)})")

    # Filter out 'N/A' values in the current categorical variable for ANCOVA and plotting
    df_subset = df_cleaned[df_cleaned[categorical_variable] != 'N/A'].copy()

    if df_subset[categorical_variable].nunique() < 2:
        print(f"Skipping {categorical_variable}: Only one unique category after filtering.")
        continue

    # Define the ANCOVA model formula
    formula = f'{dependent_variable} ~ C({categorical_variable}) + {" + ".join(continuous_covariates)}'

    # Perform ANCOVA using statsmodels
    model = ols(formula, data=df_subset).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    # Store ANCOVA results
    results = {
        'ANCOVA_F_statistic': anova_table.loc[f'C({categorical_variable})', 'F'],
        'ANCOVA_p_value': anova_table.loc[f'C({categorical_variable})', 'PR(>F)'],
    }

    # ----------------------------------------------------------------------------
    # Step 5: Check ANOVA assumptions
    # ----------------------------------------------------------------------------
    # Normality of residuals (Shapiro-Wilk test)
    shapiro_test_statistic, shapiro_p_value = stats.shapiro(model.resid)
    results['Shapiro-Wilk_statistic'] = shapiro_test_statistic
    results['Shapiro-Wilk_p_value'] = shapiro_p_value

    # Homogeneity of variances (Levene's test)
    # Prepare data for Levene's test by grouping the dependent variable by the categorical variable
    groups = [df_subset[dependent_variable][df_subset[categorical_variable] == cat].dropna() for cat in df_subset[categorical_variable].unique()]
    if len(groups) > 1:
        levene_test_statistic, levene_p_value = stats.levene(*groups)
        results['Levene_statistic'] = levene_test_statistic
        results['Levene_p_value'] = levene_p_value
    else:
        results['Levene_statistic'] = None
        results['Levene_p_value'] = None
        print(f"Skipping Levene's test for {categorical_variable}: Not enough groups.")

    analysis_results[categorical_variable] = results

    # Print ANCOVA results
    print("\nANCOVA Table:")
    print(anova_table)

    # Print Assumption Test Results
    print("\nAssumption Test Results:")
    print(f"Shapiro-Wilk Test (Normality of Residuals): Statistic={shapiro_test_statistic:.4f}, p-value={shapiro_p_value:.4f}")
    if levene_test_statistic is not None:
        print(f"Levene's Test (Homogeneity of Variances): Statistic={levene_test_statistic:.4f}, p-value={levene_p_value:.4f}")
    else:
        print("Levene's test skipped.")

    # ----------------------------------------------------------------------------
    # Step 6: Produce example boxplots
    # ----------------------------------------------------------------------------
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=categorical_variable, y=dependent_variable, data=df_subset)
    sns.pointplot(x=categorical_variable, y=dependent_variable, data=df_subset, estimator='mean', ci=95, color='red', markers='o', linestyles='--', dodge=True)
    plt.title(f'{dependent_variable} by {categorical_variable} (with Mean and 95% CI)')
    plt.xlabel(categorical_variable)
    plt.ylabel(dependent_variable)
    plt.show()

# ----------------------------------------------------------------------------
# Step 7: Summarize results in a clean table
# ----------------------------------------------------------------------------
results_df = pd.DataFrame(analysis_results).T
print("\nSummary of ANCOVA and Assumption Test Results:")
display(results_df)

# ----------------------------------------------------------------------------
# Step 8: Interpret results (Basic interpretation based on p-values)
# ----------------------------------------------------------------------------
print("\nInterpretation:")
for var, res in analysis_results.items():
    print(f"\nResults for {var}:")
    if res['ANCOVA_p_value'] < 0.05:
        print(f"  There is a statistically significant difference in {dependent_variable} across {var} groups (p={res['ANCOVA_p_value']:.4f}), controlling for age, BMI, and waist circumference.")
    else:
        print(f"  There is no statistically significant difference in {dependent_variable} across {var} groups (p={res['ANCOVA_p_value']:.4f}), controlling for age, BMI, and waist circumference.")

    if res['Shapiro-Wilk_p_value'] < 0.05:
        print(f"  Shapiro-Wilk test suggests residuals are not normally distributed (p={res['Shapiro-Wilk_p_value']:.4f}).")
    else:
        print(f"  Shapiro-Wilk test suggests residuals are normally distributed (p={res['Shapiro-Wilk_p_value']:.4f}).")

    if res['Levene_p_value'] is not None:
        if res['Levene_p_value'] < 0.05:
            print(f"  Levene's test suggests variances are not homogeneous across groups (p={res['Levene_p_value']:.4f}).")
        else:
            print(f"  Levene's test suggests variances are homogeneous across groups (p={res['Levene_p_value']:.4f}).")
    else:
        print("  Levene's test was skipped.")
```

```python
# Import necessary libraries
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
from statsmodels.stats.diagnostic import linear_rainbow, het_breuschpagan
from statsmodels.stats.outliers_influence import variance_inflation_factor
import warnings

# Ignore warnings
warnings.filterwarnings("ignore")

# ----------------------------------------------------------------------------
# Step 1 & 2: Define variables and prepare data
# ----------------------------------------------------------------------------
dependent_variable = 'DXXOFBMD'
categorical_independent_variables = ['RIAGENDR', 'PAQ605', 'PAQ650', 'ALQ101', 'SMQ020']
continuous_covariates = ['RIDAGEYR', 'BMXBMI', 'BMXWAIST']
all_predictors = categorical_independent_variables + continuous_covariates

# Ensure continuous covariates are numeric and convert 'N/A' to NaN
for col in continuous_covariates:
    df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce')

# Convert 'N/A' in categorical variables to a distinct category or NaN for consistent handling
for col in categorical_independent_variables:
    if 'N/A' in df_merged[col].unique():
        # Option 1: Convert 'N/A' to NaN (will be dropped by dropna)
        df_merged[col] = df_merged[col].replace('N/A', pd.NA)
        # Option 2: Convert 'N/A' to a distinct string category (will be included in dummy variables)
        # df_merged[col] = df_merged[col].replace('N/A', 'Unknown').astype('category')
    else:
        df_merged[col] = df_merged[col].astype('category')


# Drop rows with missing values in the dependent variable and all predictors
df_cleaned_mlr = df_merged.dropna(subset=[dependent_variable] + all_predictors).copy()

# ----------------------------------------------------------------------------
# Step 3 & 4: Convert categorical variables and Define the regression model formula
# ----------------------------------------------------------------------------
# We will use C() notation in the formula to handle categorical variables
formula = f'{dependent_variable} ~ {" + ".join([f"C({cat})" for cat in categorical_independent_variables])} + {" + ".join(continuous_covariates)}'

print("Regression formula:", formula)

# ----------------------------------------------------------------------------
# Step 5: Perform multiple linear regression
# ----------------------------------------------------------------------------
model = ols(formula, data=df_cleaned_mlr).fit()

# ----------------------------------------------------------------------------
# Step 6: Check MLR assumptions
# ----------------------------------------------------------------------------
print("\nChecking MLR Assumptions:")

# Linearity (visual inspection of residuals vs. fitted is a primary check)
# Rainbow test for linearity (can be sensitive)
try:
    rainbow_test = linear_rainbow(model)
    print(f"Rainbow Test for Linearity: Statistic={rainbow_test[0]:.4f}, p-value={rainbow_test[1]:.4f}")
except Exception as e:
    print(f"Could not perform Rainbow Test for Linearity: {e}")


# Normality of residuals
print("\nNormality of Residuals:")
# Shapiro-Wilk test
shapiro_test_statistic, shapiro_p_value = stats.shapiro(model.resid)
print(f"Shapiro-Wilk Test: Statistic={shapiro_test_statistic:.4f}, p-value={shapiro_p_value:.4f}")

# Histogram of residuals
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(model.resid, kde=True)
plt.title('Histogram of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')

# Q-Q plot of residuals
plt.subplot(1, 2, 2)
sm.qqplot(model.resid, line='s')
plt.title('Q-Q Plot of Residuals')
plt.show()

# Homoscedasticity (Constant Variance)
print("\nHomoscedasticity:")
# Breusch-Pagan test
bp_test = het_breuschpagan(model.resid, model.model.exog)
print(f"Breusch-Pagan Test: Statistic={bp_test[0]:.4f}, p-value={bp_test[1]:.4f}, F-statistic={bp_test[2]:.4f}, F-p-value={bp_test[3]:.4f}")

# Residuals vs. Fitted plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=model.fittedvalues, y=model.resid)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()

# Multicollinearity (Variance Inflation Factor - VIF)
print("\nMulticollinearity (VIF):")
# For VIF, we need the design matrix used in the model
# We need to handle categorical variables appropriately to calculate VIF
# Let's create a design matrix with dummy variables
exog = sm.tools.add_constant(model.model.exog, prepend=False)
vif_data = pd.DataFrame()
vif_data["feature"] = model.model.exog_names
vif_data["VIF"] = [variance_inflation_factor(exog, i) for i in range(exog.shape[1])]
print(vif_data)

# ----------------------------------------------------------------------------
# Step 7: Output regression summary
# ----------------------------------------------------------------------------
print("\nMultiple Linear Regression Summary:")
print(model.summary())

# ----------------------------------------------------------------------------
# Step 8: Produce plots (Predicted vs. Observed and Residuals vs. Predictors)
# ----------------------------------------------------------------------------
print("\nProducing Plots:")

# Predicted vs. Observed values
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_cleaned_mlr[dependent_variable], y=model.fittedvalues)
plt.plot([df_cleaned_mlr[dependent_variable].min(), df_cleaned_mlr[dependent_variable].max()],
         [df_cleaned_mlr[dependent_variable].min(), df_cleaned_mlr[dependent_variable].max()], color='red', linestyle='--')
plt.title('Predicted vs. Observed Values')
plt.xlabel('Observed Values')
plt.ylabel('Predicted Values')
plt.show()

# Residuals vs. Predictors plots
# Select a few predictors for plotting to avoid too many plots
predictor_plots = continuous_covariates + [categorical_independent_variables[0]] # Example: plot against continuous and one categorical

for predictor in predictor_plots:
    plt.figure(figsize=(8, 6))
    if predictor in continuous_covariates:
        sns.scatterplot(x=df_cleaned_mlr[predictor], y=model.resid)
        plt.title(f'Residuals vs. {predictor}')
    else: # Categorical variable
        sns.boxplot(x=df_cleaned_mlr[predictor], y=model.resid)
        plt.title(f'Residuals by {predictor} Categories')
    plt.axhline(y=0, color='r', linestyle='--')
    plt.xlabel(predictor)
    plt.ylabel('Residuals')
    plt.show()

# ----------------------------------------------------------------------------
# Step 9: Interpret results (Basic interpretation)
# ----------------------------------------------------------------------------
print("\nInterpretation Notes:")
print("Refer to the regression summary table for coefficients, standard errors, t-values, and p-values.")
print("The p-value for each predictor indicates its statistical significance in the model (p < 0.05 is typically considered significant).")
print("The R-squared value in the summary indicates the proportion of variance in the dependent variable explained by the predictors.")
print("\nAssumption Check Interpretation:")
print("- Normality of Residuals: Shapiro-Wilk test p-value < 0.05 suggests non-normality. Visual inspection of histogram and Q-Q plot can provide more context.")
print("- Homoscedasticity: Breusch-Pagan test p-value < 0.05 suggests heteroscedasticity (non-constant variance). The residuals vs. fitted plot can visually confirm this.")
print("- Multicollinearity: VIF values above 5 or 10 typically indicate potential multicollinearity issues.")

# ----------------------------------------------------------------------------
# Step 10: Finish task - Summary is provided in the print statements and plots
# ----------------------------------------------------------------------------
print("\nAnalysis complete. Please review the output above for the regression results, assumption checks, and plots.")
```

```python
import os
import zipfile
import pandas as pd
from glob import glob
from IPython.display import display
import statsmodels.api as sm
import statsmodels.formula.api as smf

# --- STEP 1: Extract ZIP file ---
zip_path = "/content/Biopal proj zip.zip"  # update path if needed
extract_dir = "/content/nhanes_data"

# Ensure the directory exists
os.makedirs(extract_dir, exist_ok=True)

try:
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
    print("✅ ZIP file extracted successfully.")
except FileNotFoundError:
    print(f"❌ Error: ZIP file not found at {zip_path}. Please upload the zip file to /content and try again.")
    # Removed exit() here
except Exception as e:
    print(f"❌ Error extracting ZIP file: {e}")
    # Removed exit() here


# --- STEP 2: Define variables and match files ---
file_vars = {
    'DEMO': ['SEQN', 'RIDAGEYR', 'RIAGENDR'],
    'BMX': ['SEQN', 'BMXBMI', 'BMXWAIST'],
    'PAQ': ['SEQN', 'PAQ605', 'PAQ650'],
    'SMQ': ['SEQN', 'SMQ020'],
    'ALQ': ['SEQN', 'ALQ101'],
    'BIOPRO': ['SEQN', 'LBXSCA']
}

# Collect valid .xpt files only, excluding duplicates
xpt_files = []
if os.path.exists(extract_dir): # Check if extract_dir exists before walking
    for root, dirs, files in os.walk(extract_dir):
        for file in files:
            # Check if file ends with .xpt and does not contain '(number)' like '(1)'
            if file.endswith('.xpt') and not any(char.isdigit() for char in file.split('(')[-1].split(')')[0] if '(' in file and ')' in file):
                xpt_files.append(os.path.join(root, file))

print(f"\n🔎 Found {len(xpt_files)} .xpt files after extraction.")
if xpt_files:
    print("First 5 found files:")
    for f in xpt_files[:5]:
        print(f"- {f}")


if not xpt_files:
    print("⚠️ No valid .xpt files found after extraction. Please check the contents of the zip file.")
    # Removed exit() here

# --- STEP 3: Load and merge per cycle ---
# Get unique cycles from the filenames (assuming cycle is the last part before .xpt)
all_cycles = sorted(set(os.path.basename(f).split('_')[-1].split('.')[0] for f in xpt_files if '_' in os.path.basename(f)))

print(f"\n🔄 Identified cycles: {all_cycles}")

if not all_cycles:
    print("⚠️ Could not determine cycles from filenames. Please check file naming convention.")
    # Removed exit() here


cycle_dfs = []

def load_xpt(filepath, columns):
    """Loads an XPT file and selects specified columns."""
    try:
        # Attempt to read with utf-8, fall back if needed
        try:
            df = pd.read_sas(filepath, format='xport', encoding='utf-8')
        except Exception:
            df = pd.read_sas(filepath, format='xport') # Fallback

        # Select columns that exist in the DataFrame
        return df[[col for col in columns if col in df.columns]]
    except Exception as e:
        print(f"❌ Error loading {filepath}: {e}")
        return None

if xpt_files and all_cycles: # Only proceed if files and cycles are found
    for cycle in all_cycles:
        merged = None
        print(f"\nProcessing cycle: {cycle}")
        for comp, columns in file_vars.items():
            # Find the file for the current component and cycle
            # Adjusted pattern matching to be more flexible
            matches = [f for f in xpt_files if comp.lower() in os.path.basename(f).lower() and cycle.lower() in os.path.basename(f).lower()]
            if not matches:
                print(f"  - No file found for {comp}_{cycle}")
                continue

            print(f"  - Found file for {comp}_{cycle}: {os.path.basename(matches[0])}")
            df_part = load_xpt(matches[0], columns)
            if df_part is None or df_part.empty:
                print(f"  - ⚠️ No data loaded for {comp}_{cycle}. Skipping merge for this component in this cycle.")
                continue

            if merged is None:
                merged = df_part
            else:
                # Merge on 'SEQN'
                merged = pd.merge(merged, df_part, on='SEQN', how='inner')
                print(f"  - Merged {comp}_{cycle}. Current merged shape: {merged.shape}")


        if merged is not None and not merged.empty:
            merged['Cycle'] = cycle
            cycle_dfs.append(merged)
            print(f"✅ Successfully merged data for cycle {cycle}. Added to cycle_dfs.")
        elif merged is not None and merged.empty:
             print(f"⚠️ Merged DataFrame for cycle {cycle} is empty. Skipping this cycle.")
        else:
             print(f"❌ No data merged for cycle {cycle} after checking all components.")


# --- STEP 4: Final merge ---
if cycle_dfs:
    df = pd.concat(cycle_dfs, ignore_index=True)

    # Rename columns for clarity
    df.rename(columns={
        'RIDAGEYR': 'Age',
        'RIAGENDR': 'Gender',
        'BMXBMI': 'BMI',
        'BMXWAIST': 'WaistCircumference',
        'PAQ605': 'ModerateActivity',
        'PAQ650': 'VigorousActivity',
        'SMQ020': 'SmokingStatus',
        'ALQ101': 'AlcoholUse',
        'LBXSCA': 'SerumCalcium'
    }, inplace=True)

    # Label categorical variables
    df['Gender'] = df['Gender'].map({1: 'Male', 2: 'Female'})
    df['ModerateActivity'] = df['ModerateActivity'].map({1: 'Yes', 2: 'No', 7: 'Refused', 9: "Don't know"})
    df['VigorousActivity'] = df['VigorousActivity'].map({1: 'Yes', 2: 'No', 7: 'Refused', 9: "Don't know"})
    df['SmokingStatus'] = df['SmokingStatus'].map({1: 'Yes', 2: 'No', 7: 'Refused', 9: "Don't know"})
    df['AlcoholUse'] = df['AlcoholUse'].map({1: 'Yes', 2: 'No', 7: 'Refused', 9: "Don't know"})


    # Select relevant columns
    df = df[['SEQN', 'Cycle', 'Age', 'Gender', 'BMI', 'WaistCircumference',
             'ModerateActivity', 'VigorousActivity', 'SmokingStatus',
             'AlcoholUse', 'SerumCalcium']].copy() # Create a copy to avoid SettingWithCopyWarning


    print(f"\n✅ Final merged dataset created with shape: {df.shape}\n")
    # display(df.head()) # Display head for verification
else:
    print("\n⚠️ No data was successfully merged across cycles. Check file names and structure.")
    # Removed exit() here

# --- STEP 5: Perform ANOVA tests ---
# Only proceed with ANOVA if df was successfully created
if 'df' in locals() and not df.empty:
    dependent_variable = 'SerumCalcium'
    independent_variables = ['Age', 'Gender', 'BMI', 'WaistCircumference',
                           'ModerateActivity', 'VigorousActivity', 'SmokingStatus',
                           'AlcoholUse']

    print(f"Performing ANOVA tests with '{dependent_variable}' as the dependent variable.\n")

    for var in independent_variables:
        # Drop rows with missing values for the current independent and dependent variables
        temp_df = df[[dependent_variable, var]].dropna().copy()

        if temp_df.empty:
            print(f"Skipping {var}: No data after dropping missing values.")
            continue

        # Check if the variable is categorical (object type or less than 10 unique values)
        if temp_df[var].dtype == 'object' or temp_df[var].nunique() < 10:
            # Perform One-Way ANOVA for categorical variables
            try:
                # Drop categories with less than 2 observations
                valid_categories = temp_df[var].value_counts()[temp_df[var].value_counts() >= 2].index
                temp_df = temp_df[temp_df[var].isin(valid_categories)]

                if temp_df[var].nunique() < 2:
                    print(f"Skipping {var}: Not enough categories with sufficient observations for ANOVA.")
                    continue

                formula = f'{dependent_variable} ~ C({var})'
                model = smf.ols(formula=formula, data=temp_df).fit()
                anova_table = sm.stats.anova_lm(model, typ=2) # Use Type 2 ANOVA

                print(f"📊 ANOVA Results for '{var}':")
                display(anova_table)
                print("-" * 50)

            except Exception as e:
                print(f"❌ Error performing ANOVA for {var}: {e}")

        else:
            # Perform Regression-based ANOVA for continuous variables
            try:
                formula = f'{dependent_variable} ~ {var}'
                model = smf.ols(formula=formula, data=temp_df).fit()
                anova_table = sm.stats.anova_lm(model, typ=2) # Use Type 2 ANOVA

                print(f"📊 ANOVA Results for '{var}':")
                display(anova_table)
                print("-" * 50)

            except Exception as e:
                print(f"❌ Error performing ANOVA for {var}: {e}")
else:
    print("ANOVA tests skipped because the DataFrame 'df' was not created successfully.")
```

## Interpretation of Residual Plots

The residual plots generated above show the relationship between each independent variable and the residuals (the difference between observed and predicted Serum Calcium values) from the multilinear regression model. These plots help us assess some of the key assumptions of linear regression:

*   **Linearity**: For continuous independent variables (Age, BMI, WaistCircumference), a random scatter of points around the horizontal line at zero suggests that the linear relationship assumed by the model is appropriate for that variable, after accounting for other predictors. Any curved pattern might indicate non-linearity.
*   **Homoscedasticity (Constant Variance)**: The spread of the residuals should be roughly consistent across the range of the independent variable. If the spread of residuals increases or decreases as the independent variable changes (a "cone" shape), it suggests heteroscedasticity, meaning the variance of the errors is not constant.

Based on the plots:

*   **Age**: The scatter appears relatively random around zero, but there might be some slight fanning out at higher ages, suggesting potential heteroscedasticity.
*   **Gender**: The plot shows two distinct vertical lines of residuals for Male and Female. The spread of residuals appears somewhat different between the two genders, which might suggest different variability in Serum Calcium for males and females not fully captured by the model.
*   **BMI**: There seems to be a relatively random scatter around zero across the range of BMI values, although there are some points with larger residuals, particularly at higher BMI values.
*   **WaistCircumference**: Similar to BMI, the scatter is generally random, but there are some larger residuals at higher waist circumferences.
*   **ModerateActivity**: The residuals are clustered around the 'No', 'Yes', and 'Don't know' categories. The spread of residuals appears somewhat different across these categories.
*   **VigorousActivity**: The residuals are clustered around the 'No' and 'Yes' categories. The spread seems relatively similar between the two, though there are some larger residuals.
*   **SmokingStatus**: Residuals are clustered by category ('Yes', 'No', 'Refused', "Don't know"). The spread varies across categories.
*   **AlcoholUse**: Residuals are clustered by category ('Yes', 'No', 'Refused', "Don't know"). The spread varies across categories.

**Overall Considerations**:

*   The plots for continuous variables generally show a scatter around zero, but some suggest potential issues like heteroscedasticity (Age) or larger residuals at higher values (BMI, WaistCircumference).
*   For categorical variables, the plots show clusters of residuals for each category. Differences in the spread of residuals between categories might indicate violations of the homoscedasticity assumption.
*   The presence of points far from the zero line in any plot could indicate outliers.

These visualizations provide visual clues about the model's assumptions. Formal statistical tests can also be conducted to check for heteroscedasticity (e.g., Breusch-Pagan test) and normality of residuals (e.g., Shapiro-Wilk test on the residuals), which you can consider as next steps.

```python
# Install required libraries
!pip install pandas statsmodels seaborn tabulate

import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from tabulate import tabulate

# -----------------------------
# 1. Load & Combine NHANES Data
# -----------------------------
cycles = {
    "2007_2008": "E",
    "2009_2010": "F",
    "2011_2012": "G",
    "2013_2014": "H",
    "2015_2016": "I",
    "2017_2018": "J"
}

# Dependent variable
deps = ["LBXVIDMS"]  # Serum Vitamin D

# Independent variables (categorical & numerical)
inds = ["RIDAGEYR", "RIAGENDR", "BMXBMI", "BMXWAIST", "PAQ605", "PAQ650", "ALQ101"]

# Data types to merge per cycle
data_types = ["DEMO", "BMX", "PAQ", "ALQ", "VID"]

df_all = []

# Loop through each cycle and merge relevant files
for cyc_name, cyc_suffix in cycles.items():
    cycle_dfs = []
    for dtype in data_types:
        filename = f"{dtype}_{cyc_suffix}.xpt"
        try:
            df_part = pd.read_sas(filename)
            cycle_dfs.append(df_part)
        except FileNotFoundError:
            print(f"Warning: {filename} not found. Skipping.")
    if cycle_dfs:
        df_cyc = cycle_dfs[0]
        for i in range(1, len(cycle_dfs)):
            df_cyc = pd.merge(df_cyc, cycle_dfs[i], on="SEQN", how="outer")
        available_cols = ['SEQN'] + [col for col in deps + inds if col in df_cyc.columns]
        df_cyc = df_cyc[available_cols].copy()
        df_cyc["cycle"] = cyc_name
        df_all.append(df_cyc)

# Combine all cycles
df = pd.concat(df_all, ignore_index=True)

# -----------------------------------
# 2. Data Cleaning & Type Conversion
# -----------------------------------
# Drop rows with missing dependent variable & main covariates
subset_cols_for_dropna = deps + ["RIDAGEYR", "BMXBMI", "BMXWAIST"]
df = df.dropna(subset=subset_cols_for_dropna)

# Convert selected variables to category type
cat_vars = ["RIAGENDR", "PAQ605", "PAQ650", "ALQ101"]
cat_vars_present = [v for v in cat_vars if v in df.columns]
for v in cat_vars_present:
    df[v] = df[v].astype("category")

# -----------------------------------
# 3. Function to Run ANCOVA + Checks
# -----------------------------------
def run_ancova_with_checks(dep, cat_var):
    print(f"\n===== ANCOVA: {dep} by {cat_var} (Controlling for Age, BMI, Waist) =====")

    # ANCOVA model formula
    formula = f"{dep} ~ C({cat_var}) + RIDAGEYR + BMXBMI + BMXWAIST"
    model = ols(formula, data=df).fit()

    # ANOVA table (Type II)
    anova_table = sm.stats.anova_lm(model, typ=2)
    results_table = anova_table.loc[:, ["F", "PR(>F)"]].rename(
        columns={"F": "F-value", "PR(>F)": "p-value"}
    ).round(4)

    # Add significance marking with color
    def color_p(p):
        return f"\033[92m{p:.4f}\033[0m" if p < 0.05 else f"\033[91m{p:.4f}\033[0m"

    table_data = []
    for index, row in results_table.iterrows():
        table_data.append([
            index,
            row["F-value"],
            color_p(row["p-value"])
        ])

    print("\nANOVA Results Table:")
    print(tabulate(table_data, headers=["Variable", "F-value", "p-value"], tablefmt="pretty"))

    # ---- Assumption Checks ----
    residuals = model.resid

    # Normality of residuals
    shapiro_p = stats.shapiro(residuals)[1]
    print(f"\nShapiro–Wilk Test for Normality: p = {shapiro_p:.4f} " +
          ("✅ Normal" if shapiro_p > 0.05 else "❌ Not normal"))

    # Homogeneity of variances
    group_data = [df.loc[df[cat_var] == level, dep] for level in df[cat_var].cat.categories]
    levene_p = stats.levene(*group_data, center='mean')[1]
    print(f"Levene's Test for Homogeneity: p = {levene_p:.4f} " +
          ("✅ Equal variances" if levene_p > 0.05 else "❌ Variances differ"))

    # ---- Boxplot ----
    plt.figure(figsize=(6,4))
    sns.boxplot(x=cat_var, y=dep, data=df)
    plt.title(f"{dep} by {cat_var}")
    plt.tight_layout()
    plt.show()

# -----------------------------------
# 4. Run ANCOVA for all categorical IVs
# -----------------------------------
for dep in deps:
    for cat_var in cat_vars_present:
        if df[cat_var].nunique() > 1:  # Only run if >1 group
            run_ancova_with_checks(dep, cat_var)
        else:
            print(f"\nSkipping {cat_var} — only one unique category.")
```

Gender: There’s a clear difference in vitamin D levels between men and women, even after adjusting for age, BMI, and waist size.

Physical activity (PAQ605 & PAQ650): People who reported more frequent or vigorous physical activity tended to have higher vitamin D levels. The effect was especially strong for vigorous activity (PAQ650).

Alcohol use (ALQ101): People who drank alcohol had different vitamin D levels compared to non-drinkers, even when age, BMI, and waist size were considered.

Statistical checks showed that the data didn’t follow a perfect “bell curve” (not normal) and that in most cases, group variances were different—except for alcohol use, where variance was about the same between groups.

```python
# Install required packages
!pip install pandas statsmodels seaborn matplotlib

import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy import stats

# -----------------------------
# 1. LOAD & PREPARE NHANES DATA
# -----------------------------

cycles = {
    "2007_2008": "E",
    "2009_2010": "F",
    "2011_2012": "G",
    "2013_2014": "H",
    "2015_2016": "I",
    "2017_2018": "J"
}

# Dependent and independent variables
dep_var = "LBXVIDMS"  # Vitamin D
ind_vars = ["RIDAGEYR", "RIAGENDR", "BMXBMI", "BMXWAIST", "PAQ605", "PAQ650", "ALQ101"]

data_types = ["DEMO", "BMX", "PAQ", "ALQ", "VID"]

df_all = []

# Load and merge per cycle
for cyc_name, cyc_suffix in cycles.items():
    cycle_dfs = []
    for dtype in data_types:
        filename = f"{dtype}_{cyc_suffix}.xpt"
        try:
            df_part = pd.read_sas(filename)
            cycle_dfs.append(df_part)
        except FileNotFoundError:
            print(f"Warning: {filename} not found. Skipping.")
            continue
    if cycle_dfs:
        df_cyc = cycle_dfs[0]
        for i in range(1, len(cycle_dfs)):
            df_cyc = pd.merge(df_cyc, cycle_dfs[i], on="SEQN", how="outer")
        # Keep only needed columns
        available_cols = ["SEQN", dep_var] + [col for col in ind_vars if col in df_cyc.columns]
        df_cyc = df_cyc[available_cols].copy()
        df_cyc["cycle"] = cyc_name
        df_all.append(df_cyc)

# Combine all cycles
df = pd.concat(df_all, ignore_index=True)

# --------------------------------
# 2. CLEAN DATA & CONVERT CATEGORIES
# --------------------------------

# Drop rows missing in dep_var or predictors
cols_to_check = [dep_var] + ind_vars
df = df.dropna(subset=cols_to_check)

# Convert categorical variables
cat_vars = ["RIAGENDR", "PAQ605", "PAQ650", "ALQ101", "cycle"]
for v in cat_vars:
    if v in df.columns:
        df[v] = df[v].astype("category")

# --------------------------------
# 3. MULTIPLE LINEAR REGRESSION
# --------------------------------

# Build regression formula
formula = f"{dep_var} ~ " + " + ".join([f"C({v})" if v in cat_vars else v for v in ind_vars])

# Fit model
model = ols(formula, data=df).fit()

# Print full statsmodels summary
print(model.summary())

# --------------------------------
# 4. CLEAN COEFFICIENT TABLE
# --------------------------------
coef_table = pd.DataFrame({
    "Variable": model.params.index,
    "Coefficient": model.params.values,
    "Std. Error": model.bse.values,
    "t-value": model.tvalues.values,
    "p-value": model.pvalues.values
})
print("\nClean Coefficient Table:")
print(coef_table)

# --------------------------------
# 5. CHECK ASSUMPTIONS
# --------------------------------

# Normality of residuals
shapiro_p = stats.shapiro(model.resid)[1]
print(f"\nShapiro-Wilk Test p-value (Normality): {shapiro_p:.4f}")

# Homoscedasticity (Residuals vs Fitted plot)
plt.scatter(model.fittedvalues, model.resid)
plt.axhline(y=0, color="red", linestyle="--")
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted")
plt.show()

# Residual Q-Q plot
sm.qqplot(model.resid, line='s')
plt.title("Q-Q Plot of Residuals")
plt.show()

# Multicollinearity (VIF)
X = model.model.exog
vif_df = pd.DataFrame({
    "Variable": model.model.exog_names,
    "VIF": [variance_inflation_factor(X, i) for i in range(X.shape[1])]
})
print("\nVariance Inflation Factors:")
print(vif_df)

# --------------------------------
# 6. EXAMPLE PLOT
# --------------------------------
sns.boxplot(x="RIAGENDR", y=dep_var, data=df)
plt.title(f"{dep_var} by Gender")
plt.show()
```

# Use of AI (PAIR Framework)

10.1 Problem

Our project sought to examine relationships between health behaviors and key biological markers in the U.S. population, using NHANES data. Specifically, we focused on three dependent variables: Femur Bone Mineral Density (DXXOFBMD), Serum Vitamin D (LBXVIDMS), and Serum Calcium (LBXSCA). The main question was whether lifestyle factors—such as physical activity, alcohol consumption, and smoking status—show statistically significant associations with these health outcomes after adjusting for relevant covariates like age, BMI, and waist circumference. This required statistical modeling (ANOVA, ANCOVA, and multiple linear regression) along with assumption testing and clear result visualizations.

10.2 AI

To address this, we used a combination of ChatGPT and Google Colab’s built-in Gemini AI. ChatGPT was chosen for its ability to generate clean, well-structured Python code and to explain statistical concepts in detail. It was used extensively to:

Draft and refine code for ANOVA and multiple linear regression using statsmodels

Suggest statistical assumption checks and visualizations (e.g., QQ plots, boxplots)

Help design clear, interpretable outputs with tables showing F-values, p-values, and effect sizes
Gemini AI in Google Colab was used for more in-line, iterative assistance while coding—helping troubleshoot errors, quickly testing alternative approaches, and adjusting plot formats directly in the working environment.

10.3 Interaction

Our interaction with AI was iterative and critical. We started with high-level prompts describing our dataset (df_merged with NHANES variables) and our goals (e.g., "Run ANOVA for each categorical variable controlling for age, BMI, and waist circumference"). The AI produced draft scripts, which we tested in Colab. When results were unclear, we refined prompts to request cleaner outputs, more descriptive comments, or additional statistical tests. We cross-checked outputs with NHANES codebooks to ensure proper variable interpretation, adjusted code where assumptions were not met, and combined AI-generated snippets with our own edits for efficiency and accuracy.

10.4 Reflection

Using AI accelerated our workflow by significantly reducing the time needed to code and debug statistical analyses. It also enhanced our understanding of techniques like ANOVA assumption testing and effect size interpretation. However, the AI occasionally misinterpreted variable coding (e.g., treating continuous measures as categorical), requiring manual verification. ChatGPT was stronger for structured code generation and conceptual explanations, whereas Gemini AI excelled at quick, inline adjustments in the Colab environment. Overall, AI was an effective collaborator, but human oversight was essential for ensuring data accuracy, methodological soundness, and interpretability of results.
